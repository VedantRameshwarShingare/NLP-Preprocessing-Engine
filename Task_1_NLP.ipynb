```python
# Task 2 - PREPROCESS FUNCTION
import re

def preprocess_text(text):
    
    if not text or not isinstance(text, str):
        return []

    # lowercase
    text = text.lower()

    # remove urls & emails
    text = re.sub(r"http\S+|www\S+|@\S+", "", text)

    # remove numbers
    text = re.sub(r"\d+", "", text)

    # remove repeated characters (soooo → so)
    text = re.sub(r"(.)\1{2,}", r"\1", text)

    # remove special characters / emojis
    text = re.sub(r"[^a-zA-Z\s]", "", text)

    # remove extra spaces
    text = re.sub(r"\s+", " ", text).strip()

    tokens = text.split()

    # remove short words except important ones
    tokens = [
        word for word in tokens
        if len(word) > 2 or word in ["no", "not"]
    ] 

    return tokens
```

```python
# Task 3 - STRESS TESTING
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

clean_sentences = []
all_tokens = []

for sentence in test_sentences:
    tokens = preprocess_text(sentence)
    clean_sentence = " ".join(tokens)

    clean_sentences.append(clean_sentence)
    all_tokens.extend(tokens)

    print("Original:", sentence)
    print("Tokens:", tokens)
    print("Cleaned:", clean_sentence)
    print("-" * 50)
```

```python
# Task 4 - TOKEN ANALYTICS
def token_analysis(tokens):
    if not tokens:
        return 0, 0, 0
    
    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(t) for t in tokens) / total

    return total, unique, round(avg_len, 2)


for sentence in test_sentences:
    tokens = preprocess_text(sentence)
    
    total, unique, avg = token_analysis(tokens)

    print(f"Sentence: {sentence}")
    print(f"Total Tokens: {total}")
    print(f"Unique Tokens: {unique}")
    print(f"Avg Length: {avg}")
    print("-" * 40)
```

```python
# Analysis Answers 
# Most noisy sentence:
# "Win $$$ now!!! Limited offer!!!"
# Most meaningful retained:
# "I am not happy with this"
```

```python
# Task 5 - FREQUENCY ANALYSIS
from collections import Counter

counter = Counter(all_tokens)

print("Top 10 words:")
print(counter.most_common(10))

print("\nLeast frequent words:")
print(counter.most_common()[-5:])
```

```python
def full_pipeline(text_list):
    
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens = preprocess_text(text)
        clean_sentences.append(" ".join(tokens))
        all_tokens.extend(tokens)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }


result = full_pipeline(test_sentences)
print(result)
```
